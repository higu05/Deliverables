```python
# Importing the necessary files
from model import HealthTrackerModel
from view import HealthTrackerView
from controller import HealthTrackerController

# Main Function
def main():
    # Initialize Model, View, and Controller
    model = HealthTrackerModel()
    view = HealthTrackerView()
    controller = HealthTrackerController(model, view)

    # Main Loop
    while True:
        # Display Menu
        print("\n--- Personalized Workout and Health Tracker ---")
        print("1. Enter Personal Details")
        print("2. Calculate BMI")
        print("3. Get Workout Suggestions")
        print("4. Log Activity")
        print("5. View Activity Logs")
        print("6. Exit")

        choice = view.get_user_input("Enter your choice: ")

        if choice == "1":
            controller.collect_user_data()
        elif choice == "2":
            controller.show_bmi()
        elif choice == "3":
            controller.show_workout_suggestions()
        elif choice == "4":
            controller.log_user_activity()
        elif choice == "5":
            controller.show_activity_logs()
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            view.display_message("Invalid choice. Please try again.")

# Run the main function
if __name__ == "__main__":
    main()
```
