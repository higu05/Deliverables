```python
from controller import NoteController

controller = NoteController()

def main():
    controller.load_notes_action()
    while True:
        print("\nOptions:")
        print("1. Add Note")
        print("2. Edit Note")
        print("3. Delete Note")
        print("4. Summarize Note")
        print("5. View Notes")
        print("6. Save Notes")
        print("7. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            controller.add_note_action()
        elif choice == "2":
            controller.edit_note_action()
        elif choice == "3":
            controller.delete_note_action()
        elif choice == "4":
            controller.summarize_note_action()
        elif choice == "5":
            controller.display_notes_action()
        elif choice == "6":
            controller.save_notes_action()
        elif choice == "7":
            controller.save_notes_action()
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
